```python
import gym
env = gym.make("LunarLander-v2",render_mode='human')
env.reset()
while True:
    action = int(input("Action: "))
    if action in (0, 1):
        state,reward,term,trunc,info = env.step(action)
        env.render()
        if term or trunc:
            break
         
#env.close()
        
```
